PRUEBAS DE SERVICIOS PARA LAS 3 BASES DE DATOS.

```python
import requests
import json

# Construimos la URL de la API

Host = '127.0.0.1'
Port = '5000'
Url = 'http://'+Host+':'+Port
Url
```

```python
# I. SERVICIO DE GRÁFICA DE BARRAS POR ESTADO

años = [0]
mes = [0]
entidad = [0]
municipio = [0]

data=json.dumps({"años":años,"mes":mes,"entidad":entidad,"municipio":municipio})

response = requests.post(Url+'/Acumulado/graficas/estado-barras',data=data)

# Imprimir la respuesta
print(response.text)
print("Estatus del codigo:", response.status_code)
```

```python
# II. SERVICIO DE GRÁFICA DE BARRAS POR BASE DE DATOS

años = [0]
mes = [0]
entidad = [1,2,3,4,5,6,7,8]
municipio = [0]

data=json.dumps({"años":años,"mes":mes,"entidad":entidad,"municipio":municipio})

response = requests.post(Url+'/Acumulado/graficas/porBase-barras',data=data)

# Imprimir la respuesta
print(response.text)
print("Estatus del codigo:", response.status_code)
```
