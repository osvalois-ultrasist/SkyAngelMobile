### Pruebas Reacciones (SkyONE)

#### Construir URL

```python
import requests

# Construimos la URL de la API
Host = '127.0.0.1'
Port = '5000'
Url = 'http://'+Host+':'+Port
Url
```

#### Tipo Reaccion

```python
data = {
    "tipoReaccion": 2
}

# Enviar la solicitud POST con el JSON
response = requests.post(Url+'/reacciones/municipios/delitos', json=data)

# Imprimir la respuesta
print(response.status_code)
print(response.text)
```

#### Catálogo Reacciones

```python
# Enviar la solicitud GET con el JSON
response = requests.get(Url+'/reacciones/catalogo-reacciones')

# Imprimir la respuesta
print(response.status_code)
print(response.text)
```
