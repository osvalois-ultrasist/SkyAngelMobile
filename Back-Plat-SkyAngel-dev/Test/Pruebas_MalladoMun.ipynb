# Este NB contiene pruebas para la coleccion de Mallados

```python
import requests

# Construimos la URL de la API
Host = '127.0.0.1'
Port = '5000'
Url = 'http://'+Host+':'+Port
Url
```

### ENDPOINT GEOCOMPORTAMIENTO (MALLADO)

Devuelve un mallado de hexagonos conforme a la prediccion para:
*   delito
*   fuente
*   temporalidad

```python
# Ejemplo de peticion CORRECTA                                      Tipo POST

# Definir los datos en formato JSON
data = {
    'idMunicipio': '01001',
}

# Enviar la solicitud POST con el JSON
response = requests.post(Url+'/get-malladomun', json=data)

# Imprimir la respuesta
print(response.status_code)
#print(response.json())
```

```python
# Ejemplo de peticion INCORRECTA                                      Tipo POST

# Definir los datos en formato JSON
data = {
    'idMunicipios': '01001',
}

# Enviar la solicitud POST con el JSON
response = requests.post(Url+'/get-malladomun', json=data)

# Imprimir la respuesta
print(response.status_code)
print(response.json())
```
