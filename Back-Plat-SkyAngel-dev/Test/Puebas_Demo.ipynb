# NoteBook para las pruebas de la API REST

Este notebook contiene llamados para cada uno de los endpoints de la app

```python
import requests

# Construimos la URL de la API
Host = '127.0.0.1'
Port = '5000'
Url = 'http://'+Host+':'+Port
Url
```

```python
# Prueba de conexion con Fargate
# Enviar la solicitud POST con el JSON
response = requests.get(Url+'/prueba-fargate')
print(response.status_code)
response.json()
```

```python
# Ejemplo de peticion CORRECTA                                      Tipo POST

# Definir los datos en formato JSON
data = {
    'fuente': 'secretariado',
    'ID_delito': 1,
    'size_mallado': 50,
    'temporalidad': 'LastYear'
}

# Enviar la solicitud POST con el JSON
response = requests.post(Url+'/mallados/geocomportamiento', json=data)

# Imprimir la respuesta
print(response.status_code)
print(response.json())
```

```python
import json

# Abrimos el geojson con la informacion de la ruta
Ruta = 'C:\\Users\\user\\Desktop\\Sky-Angel-Steps\\Geo_Info3\\Datos\\Ruta_test.json'

# Abrir el archivo JSON
with open(Ruta, 'r') as file:
    # Cargar el contenido del archivo en un diccionario de Python
    Ruta = json.load(file)
Ruta
```

```python
# Ejemplo de peticion CORRECTA                                      Tipo POST

# Definir los datos en formato JSON
data = {
    'fuente': 'secretariado',
    'ID_delito': 1,
    'size_mallado': 50,
    'temporalidad': 'LastYear',
    'geopuntos': Ruta
}

# Enviar la solicitud POST con el JSON
response = requests.post(Url+'/mallados/georutavalues', json=data)

# Imprimir la respuesta
print(response.status_code)
print(response.json())
```

```python
# Ejemplo de peticion CORRECTA de fecha con parametros                Tipo GET

# Definir los parmetros
params = {
    'numdia': 10,  # Ejemplo de valor para numdia
    'nummes': 7    # Ejemplo de valor para nummes
}

Respuesta = requests.get(Url+'/sumfecha',params=params)
Respuesta.json()
```
